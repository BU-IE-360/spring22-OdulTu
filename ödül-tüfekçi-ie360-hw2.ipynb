```r
hw2data <- read.csv("IE360_Spring22_HW2_data.csv",colClasses=c('character',rep('numeric',10)))
summary(hw2data)
str(hw2data)
```

```r
colnames(hw2data) <- c("Quarter", "UGS", "RNUV","NLPG","PU","PG","NUGV","NDGV","GNPA","GNPC","GNPT")
hw2data
```

```r
hw2data$Quarter=as.Date(as.yearqtr(hw2data$Quarter,format="%Y_Q%q"))
hw2data_training=data.table(hw2data[c(1:28),])
hw2data_forecast=data.table(hw2data[c(29,30,31,32),])
```

```r
str(hw2data_training)
str(hw2data_forecast)
```

```r
ggplot1 <- ggplot(data = hw2data_training,aes(x = Quarter,y = UGS))
ggplot1 + geom_line(color ="red") +
  geom_smooth(fill = NA, color="black",linetype = "twodash", size = 0.5) +
  labs(title = "UGS in Each Quarter",
       x = "Quarter",
       y = "UGS" )
```

```r
mean=roll_mean(hw2data_training$UGS,4,align='left')
plot(mean,
     type='l',col='black',
     xlab = "time",
     ylab = "rolling mean")
var=roll_var(hw2data_training$UGS,4,align='left')
plot(var,
     type='l',col='black',
     xlab = "time",
     ylab = "rolling variance"
    )
```

```r
acf(hw2data_training$UGS)
```

```r
hw2data_training[,trend := 1:.N ]
hw2data_training[,Quarter_:=as.character(month(Quarter))]
hw2data_forecast[,trend := 29:32 ]
hw2data_forecast[,Quarter_:=as.character(month(Quarter))]
model1 <- lm(UGS ~ trend+Quarter_, hw2data_training)
model1
summary(model1)
```

```r
pm1=copy(hw2data_training)
pm1[,actual:=UGS]
pm1[,pt:=predict(model1,pm1)]
pm1[,rt:=actual-pt]
ggplot(pm1 ,aes(x=Quarter)) +
        geom_line(aes(y=UGS,color='real')) + 
        geom_line(aes(y=pt,color='predicted'))
```

```r
checkresiduals(model1$residuals)
```

```r
ggpairs(hw2data_training, cardinality_threshold=30)
```

```r
hw2data_training$UGSlag1=lag(hw2data_training$UGS,1)
hw2data_training$NUGVlag5=lag(hw2data_training$NUGV,5)
hw2data_training$NDGVlag5=lag(hw2data_training$NDGV,5)
model2 <-  lm(UGS ~ trend+PU+PG+NUGV+NDGV+GNPA +UGSlag1+NUGVlag5+NDGVlag5 + Quarter_, hw2data_training)
summary(model2)
```

```r
checkresiduals(model2$residuals)
```

```r
pm2=copy(hw2data_training)
pm2[,actual:=UGS]
pm2[,pt:=predict(model2,pm2)]
pm2[,rt:=actual-pt]
ggplot(pm2 ,aes(x=Quarter)) +
        geom_line(aes(y=UGS,color='real')) + 
        geom_line(aes(y=pt,color='predicted'))
```

```r
model3= lm(UGS ~ trend+NLPG+PU+PG+NUGV+NDGV+GNPA + Quarter_, hw2data_training)
summary(model3)
```
